# Carga de datos

````python
import numpy as np
import random
import math
import matplotlib.pyplot as plt
import GA
from GA import *
````

````python
with open("berlin52.tsp", "r") as file:
    while True:
        line = file.readline()
        if line == "NODE_COORD_SECTION\n":
            break

    data = []

    while True:
        line = file.readline()
        if line == "EOF\n":
            break

        str_array = line.split()
        data.append(
            (float(str_array[1]), float(str_array[2]), int(str_array[0])))
````

````python
with open("berlin52.opt.tour", "r") as file:
    while True:
        line = file.readline()
        if line == "TOUR_SECTION\n":
            break

    OPTIMAL = []

    while True:
        line = file.readline()
        if line == "-1\n":
            break
        OPTIMAL.append(int(line))
````

````python
data_array = np.array(data)
OPTIMAL_array = np.array(OPTIMAL)
camino_optimo = data_array[OPTIMAL_array - 1]
````

# Funciones útiles

````python
def plot_path(best_individual, title: str = None):
    X = [city[0] for city in best_individual]
    Y = [city[1] for city in best_individual]
    city_numbers = [city[2] for city in best_individual]

    # Lista de colores para cada punto
    colors = [0] + [1] * (len(best_individual) - 2) + [2]

    plt.figure(figsize=(16, 9))
    if title is not None:
        plt.title(title)

    #Pintar lineas
    plt.plot(X, Y)

    #Línea entre el primer y último punto en rojo
    last_line_X = [X[len(X) - 1], X[0]]
    last_line_Y = [Y[len(Y) - 1], Y[0]]
    plt.plot(last_line_X, last_line_Y, color="red")

    # Pintar puntos
    plt.scatter(X, Y, c=colors)
    plt.show()
    print("Distancia recorrida:", euclidean(best_individual))


def plot_and_return_n_ga_results(ga, n=1):
    results = []
    best_individuals = []
    for i in range(n):
        best_individual, best_fitness, last_population = ga.solve(data)
        plt.xlabel('generation')
        plt.ylabel('longitud del camino')
        plt.plot(best_fitness)
        plt.show()
        fitness = ga.fitness_function(best_individual)
        print("Distancia recorrida:", fitness)
        best_individuals.append(best_individual)
        results.append(fitness)
    return {'results': results, 'best_individuals': best_individuals}
````

## Funciones ideadas por el equipo
### Random mutate
Esta función se aplica una vez por cada gen del individuo sin ninguna restricción.
```python
def mutate(individual, mutationRate, **params):
    for swapped in range(len(individual)):
        if (random.random() < mutationRate):
            swapWith = int(random.random() * len(individual))

            city1 = individual[swapped]
            city2 = individual[swapWith]

            individual[swapped] = city2
            individual[swapWith] = city1
    return individual
```

````python
ga_mutate = genetic_algorithm_class(crossover=crossover_two_points,
                                    mutate=mutate,
                                    selection_method=select_by_tournament,
                                    fitness_function=euclidean,
                                    population_size=30,
                                    elite_size=1,
                                    tournament_proportion=0.2,
                                    generations=6000,
                                    mutation_rate=0.005)
````

````python
res_mutate = plot_and_return_n_ga_results(ga_mutate)
````

````python
for ind in res_mutate['best_individuals']:
    plot_path(ind)
````

### Mutate_max_range
Esta función se aplica una vez por cada gen del individuo con la restricción de que solo se podrá intercambiar con una ciudad que esté a la distancia permitida por el parámetro *max_range*.
```python
def mutate_max_range(individual, mutationRate, **params):
    for swapped in range(len(individual)):
        if (random.random() < mutationRate):
            chosen = int(random.random() * params['range_of_mutation'])
            if random.random() < 0.5:
                swapWith = (swapped - chosen) % len(individual)
            else:
                swapWith = (swapped + chosen) % len(individual)

            city1 = individual[swapped]
            city2 = individual[swapWith]

            individual[swapped] = city2
            individual[swapWith] = city1
    return individual
```

````python
ga_max_range = genetic_algorithm_class(crossover=crossover_two_points,
                                       mutate=mutate_max_range,
                                       selection_method=select_by_tournament,
                                       fitness_function=euclidean,
                                       population_size=30,
                                       elite_size=1,
                                       tournament_proportion=0.2,
                                       generations=6000,
                                       mutation_rate=0.005,
                                       range_of_mutation=15)
````

````python
res_max_range = plot_and_return_n_ga_results(ga_max_range)
````

````python
for ind in res_max_range['best_individuals']:
    plot_path(ind)
````

### OPT-2
```python
def mutate_opt_2(individual, mutationRate, **params):
    if random.random() > mutationRate:
        return individual

    r1 = random.randint(0, len(individual)-1)
    while True:
        r2 = random.randint(0, len(individual)-1)
        if r1 != r2:
            break
    return individual[:r1] + list(reversed(individual[r1:r2])) + individual[r2:]
```

### OPT-3
```python
def mutate_opt_3(individual, mutationRate, **params):
    if random.random() > mutationRate:
        return individual

    while True:
        i = random.randint(0, len(individual)-3)
        j = random.randint(i+1, len(individual)-2)
        k = random.randint(j+1, len(individual)-1)
        if i != j & j != k & k != i:
            break
    new_individual = individual.copy()
    A, B, C, D, E, F = new_individual[i-1], new_individual[i], new_individual[j -
                                                                              1], new_individual[j], new_individual[k-1], new_individual[k % len(new_individual)]
    d0 = distance(A, B) + distance(C, D) + distance(E, F)
    d1 = distance(A, C) + distance(B, D) + distance(E, F)
    d2 = distance(A, B) + distance(C, E) + distance(D, F)
    d3 = distance(A, D) + distance(E, B) + distance(C, F)
    d4 = distance(F, B) + distance(C, D) + distance(E, A)

    if d0 > d1:
        new_individual[i:j] = reversed(new_individual[i:j])
        return new_individual
    elif d0 > d2:
        new_individual[j:k] = reversed(new_individual[j:k])
        return new_individual
    elif d0 > d4:
        new_individual[i:k] = reversed(new_individual[i:k])
        return new_individual
    elif d0 > d3:
        tmp = new_individual[j:k] + new_individual[i:j]
        new_individual[i:k] = tmp
        return new_individual

    return individual
```

````python
res_opt_3 = plot_and_return_n_ga_results(ga_opt_3)
````
