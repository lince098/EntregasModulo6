En este trabajo se persigue la consecución de un modelo ML que sea capaz de clasificar diferentes cardiotocografias de fetos en diferentes estados. Estos estados(etiquetas) serían: "sueño tranquilo", "sueño REM", "vigilancia tranquila", "vigilancia activa", "patrón de turno (A o Susp con turnos)", "patrón acelerativo/decelerativo (situación de estrés)", "patrón decelerativo (estimulación vagal)", "patrón mayormente decelerativo", "patrón plano-sinusoidal (estado patológico)", "patrón sospechoso".

Como se observa, se tienen 10 etiquetas, por lo que se deben de usar modelos multiclases. El proceso que se va a seguir es el entrenamiento de diversos modelos de clasificación multiclases que se encuentran en la libreria scklearn. Para evitar que estos entrenamientos sean tediosos, se va a hacer uso de piplines y multiprocesamiento. 

Los modelos a utilizar son los siguientes:

* LinearDiscriminantAnalysis: es un modelo útil para clasificar datos en dos o más clases al encontrar un vector de proyección que maximiza la separación entre las clases, asumiendo que las características tienen una distribución normal y que las clases tienen la misma varianza.

* Árboles de decisión: se utilizan para problemas de clasificación y también para problemas de regresión. Son fáciles de interpretar y explicar, y pueden manejar características numéricas y categóricas.

* Random Forest: es un conjunto de árboles de decisión que se utilizan para problemas de clasificación y también para problemas de regresión. Pueden manejar características numéricas y categóricas, y son robustos a valores atípicos y ruido en los datos.

* Support Vector Machines (SVM): se utilizan para problemas de clasificación binaria o multiclase y también para problemas de regresión. SVM encuentra la mejor separación lineal o no lineal entre las diferentes clases.

* K vecinos más cercanos (KNN): se utiliza para problemas de clasificación y regresión. Busca los k vecinos más cercanos en el espacio de características y utiliza su etiqueta para hacer una predicción.

* Gradient Boosting: es un conjunto de árboles de decisión que se utilizan para problemas de clasificación y también para problemas de regresión. Se construye iterativamente y busca minimizar el error de clasificación en cada iteración.

* GaussianNB: es un algoritmo de clasificación probabilístico que se basa en el teorema de Bayes y la suposición "ingenua" de que todas las características son independientes entre sí y tienen una distribución normal (gaussiana) dentro de cada clase. 

* NuSVC es un modelo de clasificación de vectores de soporte (SVM) no lineal que utiliza una estrategia de optimización diferente a la del modelo SVM estándar. A diferencia del modelo SVM estándar, que utiliza un parámetro de regularización C para controlar la complejidad del modelo, el modelo NuSVC utiliza un parámetro de regularización diferente llamado nu. Este modelo es especialmente útil en problemas de clasificación no lineales, donde el límite de decisión entre las clases no se puede expresar de manera lineal en el espacio de características original.

* Logistic Regression: La regresión logística es un modelo de aprendizaje supervisado utilizado para la clasificación binaria o multiclase. Se utiliza para predecir la probabilidad de una variable categórica dependiente en función de una o más variables independientes continuas o categóricas. El modelo utiliza una función sigmoide para convertir una entrada lineal en una salida no lineal entre 0 y 1, que representa la probabilidad de pertenecer a cada clase.

#### Se observan gran cantidad de outliers. Eliminarlos sería una opción, aunque no la más acertada, ya que estos outliers son portadores de informacion importante para la clasificación de los fetos. Es preferible probar modelos robustos contra los outliers que empezar a eliminar información del conjunto de datos. Por ello se debe de realizar una elección critica de los modelos. Por lógica, todos los modelos donde la decisión sobre la clase se basen en métricas o en optimización de funciones de error, van a tener un rendimiento muy afectado por los valores atípicos. Siguiendo este razonamiento, se deduce que modelos como regression(basado en función exponencial), KNN(distancia entre puntos) o máquinas de vectores de soporte, que buscan maximizar las distancia al hiperplano entre los valores más cercanos de diferentes clases, van a tener un rendimiento muy lejos del óptimo. Por otra parte, tampoco resulta lógica utilizar un modelo de Naive Bayes, ya que cuando el modelo detecte un valor atípico, le va a dar la probabilidad de 0, y luego al multiplicarse las probabilidades para calcular la probabilidad de la clase, esa probabilidad de 0 va a ser un problema. Lo más acertado en este caso sería trabajar con arboles de decisión y realizar un entrenamiento de diversos tipos de modelos basados en estos árboles. El motivo es por que los árboles de decisión se basan en un criterio de partición y no en distancias entre puntos o la minimización de funciones de pérdidas, por lo que son mucho mas resistentes a outliers.

#### Se van a realizar diferentes reducciones de dimensiones mediante PCA, esto ayuda a disminuir la complejidad del dataset, por lo tanto puede mejorar la precisión de los modelos, reducir el tiempo de entrenamiento. Como tenemos diferentes modelos y a priori no se puede saber cual reducción representa mejor la variabilidad del dataset, se va a buscar el óptimo.

#### Ahora se va a crear una lista de pipelines que recoga los procedimientos para cada modelo con los que se va a trabajar, estos modelos son DecissionTreeClasifier, RandomForestClassifier, AdaBoostClassifier, GradientBoostingClassifier.

```python
pipelines = []
pipelines.append(('dt', Pipeline([('pca', PCA()), ('dt', DecisionTreeClassifier())])))
pipelines.append(('rf', Pipeline([('pca', PCA()), ('rf', RandomForestClassifier())])))
pipelines.append(('ada', Pipeline([('pca', PCA()), ('ada', AdaBoostClassifier())])))
pipelines.append(('gb', Pipeline([('pca', PCA()), ('gb', GradientBoostingClassifier())])))

param_grid = {}
```

#### Se realiza la definición de los parámetros para la búsqueda por grilla en los diferentes modelos. Los parámetros a ensayar en el PCA se van a dejar en 5,10 y 15, ya que se tiene 23 columnas. Se han elegido los siguientes modelos para entrenar con sus diferentes motivos:


#### Se va a proceder a ajustar el modelo  DecisionTreeClassifier():

* criterion: especifica la función que se utilizará para medir la calidad de una división. Puede ser "gini" para la impureza de Gini o "entropy" para la ganancia de información.

* max_depth: especifica la profundidad máxima del árbol. Si no se especifica o se establece en None, el árbol se expandirá hasta que todas las hojas contengan menos de min_samples_split ejemplos. Como el modelo se va a mover entre 5,10 o 15 columnas, se van a probar los siguiente valores: [10,20,30,40,50,60].

* min_samples_split: especifica el número mínimo de ejemplos requeridos para dividir un nodo interno. Si un nodo tiene menos de min_samples_split ejemplos, no se dividirá. Como se tienen muchs clases, se deja entre [2,5,10].

* min_samples_leaf: especifica el número mínimo de ejemplos requeridos para estar en una hoja. Si una hoja tiene menos de min_samples_leaf ejemplos, se fusionará con su nodo hermano. Los dejaremos por defecto, cuyo valor es 1.

```python
param_grid['dt'] = {'pca__n_components': [5, 10, 15],'dt__criterion':['gini','entropy'],'dt__max_depth': [10,20,30,40,50,60],
                 'dt__min_samples_split': [2, 5, 10]}
```

#### Se va a proceder a ajustar el modelo RandomForestClassifier():

* n_estimators: el número de árboles en el bosque. Se dejará por defecto(100)

Nota: Los demás parámetros son iguales que en el ejemplo anterior.


```python
param_grid['rf'] =  {'pca__n_components': [5, 10, 15,20],'rf__criterion':['gini','entropy'],'rf__max_depth': [20,30,40,50,60],
                 'rf__min_samples_split': [2, 5, 10]}
```

```python
# FALTA POR ESTUDIAR ESTOS PARÁMETROS

param_grid['lda'] = {'pca__n_components': [5, 10, 15], 'solver': ['svd', 'lsqr', 'eigen'], 'shrinkage':[None,'auto','float']}ç
param_grid['lr'] = {'pca__n_components': [5, 10, 15], 'C': [0.5,1,1.5],
                 'penalty':['l1','l2','elasticnet'], 'solver': ['newton-cg', 'lbfgs', 'sag','saga']}
```

```python
from joblib import Parallel, delayed

# Se define una función para ejecutar la búsqueda de grid para una pipeline dada
def search_pipeline(pipeline, param_grid):
    grid_search = GridSearchCV(pipeline, param_grid=param_grid, cv=5)
    grid_search.fit(X, y)
    return grid_search

# Se crea una lista de trabajos a ejecutar
jobs = [delayed(search_pipeline)(pipeline, param_grid[name]) for name, pipeline in pipelines]

# Se ejecuta los trabajos en paralelo
results = Parallel(n_jobs=-1)(jobs)

# Se imprime los resultados
for i, (name, pipeline) in enumerate(pipelines):
    print(f"Mejor modelo para {name}: {results[i].best_params_}")
    print(f"Precisión media en validación cruzada: {results[i].best_score_}\n")
```
